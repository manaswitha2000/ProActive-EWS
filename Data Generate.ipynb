# Generating dummy data 

## for spo2

```python
import random
id = []
col = []
label = []
#Range of spo2 is given here
for i in range(0,500):
    id.append(i)
    spo2 = random.randint(85,100)
    col.append(spo2)
    if(spo2 > 97):
        label.append("No")
    elif(spo2 > 95):
        label.append("Mild")
    elif(spo2 > 90):
        label.append("Moderate")
    else:
        label.append("Severe")
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
data = { 'ID': id,
        'Spo2': col,
        'Label': label}
df = pd.DataFrame(data)
```

```python
df.to_csv('sleep_data_time.csv',index = False)
```

## for temperature

```python
import random
id = []
col = []
label = []
#Range of spo2 is given here
for i in range(0,500):
    id.append(i)
    temp = random.uniform(35,38);# generates random float value 
    col.append(temp)
    if(temp > 36.1 and temp < 37.2):
        label.append("Normal")
    else:
        label.append("Abnormal")
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
data = { 'ID': id,
        'Temp': col,
        'Label': label}
df = pd.DataFrame(data)
```

```python
df.to_csv('temp_data_time.csv',index = False)
```
